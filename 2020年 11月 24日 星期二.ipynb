```python
labels = ['Full', 'Associate', 'Assistant']
men_means = [98.1, 72.7, 61.5]
women_means = [110.5, 74.9, 61.6]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars
```

```python
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='public')
rects2 = ax.bar(x + width/2, women_means, width, label='private')

ax.set_ylabel('Average Salary')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
