```python
y = []
for i in range(len(x1)):
    if x1[i] >= -0.2 and x1[i] <= 0.2:
        y.append(1)
    else:
        y.append(0)
print(y)
x = x1
plt.bar(x,y,width=1)
plt.xlim(left=-0.5,right=0.5)
```

```python
x = np.linspace(0,20,100)
y = 0.2*np.e**(-2*x)
plt.xlim(0,20)
plt.plot(x,y)
plt.grid()
```

```python
def f(x,mu=5,sigma=1):
    a = -(x-mu)**2 / (2 * sigma**2)
    b = np.e**a
    c = 1/sigma*np.sqrt(2*np.pi)
    f = c*b
    return f
```

```python
x = np.linspace(0,10,100)
plt.plot(x,f(x,4,0.5))
plt.xticks(range(11))
plt.show()
```
