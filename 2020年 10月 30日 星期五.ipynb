```python
import numpy as np
import matplotlib.pyplot as plt

x0 = 0
y0 = 1
n = 4
deltax = 0.1
x = np.linspace(0,10,n)
f1 = x*x

def f(x,y):
    return 1+y
y = np.zeros([n])
y[0] = y0

for i in range(1,n):
    y[i] = y0 + f(x[i-1],y0) * deltax
    y0 = y[i]

print("x_n\t        y_n")
for i in range(n):
    print(format(x[i],'6f'),"\t",format(y[i],'6f'))

fig, ax = plt.subplots()
ax.plot(x,y,'bx',label='linear')
ax.plot(x,f1,'r-',label='quadratic')
plt.xlabel("Value of x")
plt.ylabel("Value of y")
plt.title("Approximation Solution with Euler's Method")
ax.legend()
plt.show()
```

```python
def multmatrix(a,b):
    #Returns the product of matrix a and matrix b
    m = len(a) #number of rows in first matrix
    n = len(b[0]) #number of columns in second matrix
    newmatrix = []
    for i in range(m):
        row = []
        #for every column in b
        for j in range(n):
            sum1 = 0
            #for every element in the colum
            for k in range(len(b)):
                sum1 += a[i][k]*b[k][j]
            row.append(sum1)
        newmatrix.append(row)
    return newmatrix
```

```python
a = [[1,2,-3,-1]]
b = [[4,-1],
     [-2,3],
     [6,-3],
     [1,0]]
print(multmatrix(a,b))
```

```python
def transpose(a):
    '''Transposes matrix a'''
    output = []
    m = len(a)
    n = len(a[0])
    #create an n x m matrix
    for i in range(n):
        output.append([])
        for j in range(m):
            #replace a[i][j] with a[j][i]
            output[i].append(a[j][i])
    return output
```

```python
a = [[1,2,-3,-1]]
transpose(a)
```

```python
from math import sqrt
def quad(a,b,c): 
    '''Returns the solutions of an equation
    of the form a*x**2 + b*x + c = 0'''
    x1 = (-b + sqrt(b**2 - 4*a*c))/(2*a)
    x2 = (-b - sqrt(b**2 - 4*a*c))/(2*a)
    return x1,x2
```

```python
quad(1,3,-10)
```
